```python
from numpy.random import *
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import collections
```

```python
class Randnplot:
    def __init__(self,mu,var):                  # コンストラクタ
        self.mu = mu
        self.var = var
        print("class created")
    
    def plotdata(self, list):
        plt.hist(list,normed=True)
        x1 = np.arange(-5,5,0.01)
        y1 = np.exp(-(x1**2)/2)/np.sqrt(2 * np.pi)
        plt.plot(x1,y1)
        plt.show()
    
    def makepoisongraph(self,n):
        list = []
        for i in range(1000):
            x = 0
            for i in range(n):
                x += (poisson(lam=self.mu)-self.mu)/(np.sqrt(n*self.mu))
            list.append(x)
        self.plotdata(list)
    def makesnormalgraph(self,n):
        list = []
        for i in range(1000):
            x = 0
            for i in range(n):
                x += (normal(self.mu,self.var)-self.mu)/(np.sqrt(n*(self.var)**2))
            list.append(x)
        self.plotdata(list)
    def makesbetagraph(self,n):
        list = []
        for i in range(1000):
            x = 0
            for i in range(n):
                x += (beta(self.mu,self.var)-self.mu)/(np.sqrt(n*(self.var)**2))
            list.append(x)
        self.plotdata(list)
```

```python
graph = Randnplot(3,3)
graph.makepoisongraph(100)
```

```python
graph.makesnormalgraph(10000)
```
